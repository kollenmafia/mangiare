```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.matprat.no/oppskrifter/familien/sandwich-med-grillet-orret/'
page = requests.get(url)
soup = bs(page.content, "html.parser")
```

```python
results = soup.find(id = "artikkelen-overskrift")
item_elements = results.find_all("div", class_ = "ingredientsList")
```

```python
for item in item_elements:
    name = item.find('li', class_ = 'ingredients')
    print(name)
```
