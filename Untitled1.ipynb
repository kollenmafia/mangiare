```python
text = []
```

```python
url = "https://www.godt.no/oppskrift/8608/dampede-blaaskjell-med-hvitloek"
options = Options()
options.add_argument("start-maximized")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get(url)

ele = driver.find_element(By.ID, "application")

text.append(ele.text)


print(driver.title)
print(ele.text)


driver.close()
```

```python
df = pd.DataFrame(text)
```

```python
df.to_excel("recipee2.xlsx")
```
